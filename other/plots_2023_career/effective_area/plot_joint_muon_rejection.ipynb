```python
import numpy as np #import numpy
import matplotlib.pyplot as plt 
import matplotlib
from matplotlib.pyplot import rcParams
rcParams['mathtext.default'] = 'regular'
import math
from pylab import setp
import helper
```

```python
# stoch hist stuff
data = np.load('ehenextgen_stoch_weights.npz')


# timing data
# https://user-web.icecube.wisc.edu/~mamin/cos/200120_133644_43767651/200120_133644_43767651.html
cr = np.genfromtxt('cr_candidate.csv',delimiter=',', skip_header=1,names=['dist','time'])
# https://user-web.icecube.wisc.edu/~mamin/nu/200926_134533_53384881/200926_134533_53384881.html
nu = np.genfromtxt('nu_candidate.csv',delimiter=',', skip_header=1,names=['dist','time'])
```

```python
plt.rc('text',usetex=True) #plot with Latex
plt.rc('font',family='serif') #Used a serif font
plt.rc('font', weight='heavy') #Make the text bold\

# plt.rc('font',**{'family':'serif'})

fig2 = plt.figure(figsize=(8.5,3.5))
ax2_stoch = fig2.add_subplot(1,2,1)
ax2_timing = fig2.add_subplot(1,2,2)

########
# stoch figure
########
plot_bins = np.linspace(2, 14, 51)
n, b, p = ax2_stoch.hist(
        data['cor_values'], weights=data['cor_weights'],
        bins=plot_bins,
        histtype='step', linewidth=2, linestyle='--',
        label='Muon Bundles', density=True
)
n, b, p = ax2_stoch.hist(
        data['numu_values'], weights=data['numu_values'],
        bins=plot_bins,
        histtype='step', linewidth=2,
        label='Single High \n Energy Tracks', density=True
)
sizer=13
ax2_stoch.set_xlim([3,13])
ax2_stoch.set_xlabel('Stochasticity',size=sizer)
ax2_stoch.set_ylabel('Normalized Density',size=sizer)
ax2_stoch.tick_params(labelsize=sizer)
this_legend= ax2_stoch.legend(loc='upper right')
setp(this_legend.get_texts(), fontsize=10)
setp(this_legend.get_title(), fontsize=10)
this_legend.get_frame().set_edgecolor('k')
ax2_stoch.xaxis.set_ticks([4, 8, 12])
ax2_stoch.set_title("Shape Discrimination",size=sizer+2)

########
# timing figure
########

ax2_timing.plot(cr['dist'],cr['time'],'o',markersize=4, label='Muon Bundle \n (IC200120A)')
ax2_timing.plot(nu['dist'],nu['time'],'s',markersize=4, label='Single High\nEnergy Track\n(IC200926A)')
ax2_timing.set_xlim([0,100])
ax2_timing.set_ylim([-100,100])


ax2_timing.set_xlabel('DOM to Track Distance [m]',size=sizer)
ax2_timing.set_ylabel('Time Residual [ns]',size=sizer)
ax2_timing.tick_params(labelsize=sizer)
this_legend= ax2_timing.legend(loc='lower right',framealpha=1,handletextpad=0.1)
setp(this_legend.get_texts(), fontsize=10)
setp(this_legend.get_title(), fontsize=10)
this_legend.get_frame().set_edgecolor('k')
ax2_timing.set_title("Timing Discrimination",size=sizer+2)

# for a in [ax]:
#     a.yaxis.label.set_fontsize(18)
#     a.xaxis.label.set_fontsize(18)
#     a.tick_params(axis='both', which='major', labelsize=15)


plt.tight_layout()
plt.subplots_adjust(wspace=0.45)
fig2.savefig("combo_muon_rejection.pdf",edgecolor='none',bbox_inches="tight",dpi=300) #save the figure    
```
